#### Now that our data types have been converted, we can conduct a preliminary high-level analysis of our MLS dataset.

#### First, we can examine Toronto and see the year-over-year changes by property type.

```python
# Create a new data frame to isolate the data for the entire City of Toronto.

df_hpi_toronto = df_hpi.loc[df_hpi['Toronto Municipality'] == 'City of Toronto'].copy()
df_hpi_toronto['Year'] = df_hpi_toronto['Effective Date'].dt.year

# Group the columns for the property type.

group_cols = [
    'HPI - Single-Family Detached',
    'HPI - Single-Family Attached',
    'HPI - Townhouse',
    'HPI - Apartment']


# Calculate the average HPI by property type and year.

df_hpi_toronto_by_year = (df_hpi_toronto.groupby('Year', as_index=False)[group_cols].mean())
```

```python
df_hpi_toronto
```

```python
df_hpi_toronto_by_year
```

```python
# Now that we have the average HPI by property type by year, we can plot this on a line graph.

hpi_line = df_hpi_toronto_by_year.plot(x='Year', y=group_cols,marker='o', figsize=(16, 9))

# Format chart labels.

plt.title('Average HPI by Types by Year for the City of Toronto', fontsize=15)
plt.xlabel('Year',fontsize=15)
plt.ylabel('HPI Value',fontsize=15)
hpi_line.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))

# Add and format data labels.

for line in hpi_line.get_lines():
    x_vals = line.get_xdata()
    y_vals = line.get_ydata()
    for x, y in zip(x_vals, y_vals):
        hpi_line.annotate(
            f'${y:,.0f}',
            xy=(x, y),   
            xytext=(5, 5),
            textcoords='offset points',
            ha='center')
plt.show()
```

#### From the line graph above, we can see a clear pattern of HPI values increasing YoY.

#### The average values are vastly different by property type, so it would be good to understand the YoY value difference and percentages by property type.

```python
# Create a new data frame that calculates the HPI value difference YoY.

df_hpi_toronto_diff = df_hpi_toronto_by_year[group_cols].diff()
```

```python
df_hpi_toronto_diff
```

```python
# Calculate the average YoY value difference for each HPI property type.

toronto_avg_diff = df_hpi_toronto_diff.mean()
```

```python
# Create a bar chart to showcase the average HPI difference by property type in descending order.

toronto_avg_diff_sorted = toronto_avg_diff.sort_values(ascending=False)

plt.figure(figsize=(16, 9))
bars = plt.bar(
    toronto_avg_diff_sorted.index, 
    toronto_avg_diff_sorted.values)

# Add data labels.

for bar in bars:
    height = bar.get_height()
    label_text = f"${height:,.0f}"
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height, 
        label_text,
        ha="center", 
        va="bottom", fontsize=12)

# Set titles and formatting.

avg_diff = plt.gca()
avg_diff.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
plt.title("Average YoY Increase by HPI Type (2015-2024)", fontsize=15)
plt.xlabel("HPI Type", fontsize=15)
plt.ylabel("Average YoY Increase Value", fontsize=15)
plt.xticks(rotation=0)
plt.show()
```

```python
# Create a new data frame that calculates the HPI value % difference YoY.

df_hpi_toronto_pct_diff = df_hpi_toronto_by_year[group_cols].pct_change()*100
```

```python
df_hpi_toronto_pct_diff
```

```python
# Calculate the average YoY percent difference for each HPI property type.

toronto_avg_pct_diff = df_hpi_toronto_pct_diff.mean()
```

```python
# Create a bar chart to showcase the average HPI % difference by property type in descending order.

toronto_avg_pct_diff_sorted = toronto_avg_pct_diff.sort_values(ascending=False)

plt.figure(figsize=(16, 9))
bars = plt.bar(
    toronto_avg_pct_diff_sorted.index, 
    toronto_avg_pct_diff_sorted.values)

# Add data labels.

for bar in bars:
    height = bar.get_height()
    label_text = f"{height:.1f}%"  # Format as percentage with one decimal place
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height, 
        label_text,
        ha="center", 
        va="bottom", fontsize=15)

# Set titles and formatting.
avg_perc_diff = plt.gca()
avg_perc_diff.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.title("Average YoY Percentage Increase by HPI Type (2015-2024)", fontsize=15)
plt.xlabel("HPI Type", fontsize=15)
plt.ylabel("Average YoY % Increase", fontsize=15)
plt.xticks(rotation=0)
plt.show()
```

#### While the HPI value increase year over year differs, the percentage increase is about the same, at 8%, regardless of property type.  However, it is still interesting to note that Apartments have a better rate of return while having the lowest property value.

#### The next step is to understand the YoY property value difference and percentage by neighbourhood to determine which neighbourhoods yield better returns.

```python
# Check to see the count of unique neighbourhoods in Toronto.

df_hpi['Toronto Municipality'].value_counts()
```

```python
# Create a new data frame to showcase only the specific neighbourhoods in Toronto.

df_hpi_neighbourhood = df_hpi.loc[df_hpi['Toronto Municipality'] != 'City of Toronto'].copy()
df_hpi_neighbourhood['Year'] = df_hpi_neighbourhood['Effective Date'].dt.year


# Calculate the average HPI by neighbourhood, property type and year.
df_hpi_neighbourhood_by_year = (df_hpi_neighbourhood.groupby('Year', as_index=False)[group_cols].mean())
```

```python
df_hpi_neighbourhood
```

```python
# Create a new data frame to group the YoY value difference by neighbourhood, property type and year.

group_cols = [
    'HPI - Single-Family Detached',
    'HPI - Single-Family Attached',
    'HPI - Townhouse',
    'HPI - Apartment']

df_grouped = (df_hpi_neighbourhood.groupby(['Toronto Municipality', 'Year'], as_index=False)[group_cols].mean())

# Sort the data by municipality and year.

df_grouped.sort_values(['Toronto Municipality', 'Year'], inplace=True)

# Calculate YoY difference and percent change (by municipality)
df_diff_neighbourhood = df_grouped.groupby('Toronto Municipality')[group_cols].diff()
df_diff_neighbourhood.columns = [f"{col} YoY Diff" for col in df_diff_neighbourhood.columns]

df_pct_neighbourhood = df_grouped.groupby('Toronto Municipality')[group_cols].pct_change() * 100
df_pct_neighbourhood.columns = [f"{col} YoY %" for col in df_pct_neighbourhood.columns]

# Combine all three data frames
df_final = pd.concat([df_grouped, df_diff_neighbourhood, df_pct_neighbourhood], axis=1)
```

```python
df_final
```

#### Now we can check for each neighbourhood's YoY value difference.

```python
# Group the property type columns.

diff_cols = [
    "HPI - Single-Family Detached YoY Diff",
    "HPI - Single-Family Attached YoY Diff",
    "HPI - Townhouse YoY Diff",
    "HPI - Apartment YoY Diff"]

# Get the average YoY value difference for all property types by neighbourhood.

df_diff_by_muni = (
    df_final
    .groupby("Toronto Municipality")[diff_cols]
    .mean()                       # => averages each HPI type's YoY Diff over all years
    .mean(axis=1)                # => average across the 4 HPI types
    .rename("avg_diff")          # => name this new column "avg_diff"
    .reset_index()               # => make "Toronto Municipality" into a normal column
)

# Sort the data in descending order.

df_diff_by_muni.sort_values("avg_diff", ascending=False, inplace=True)
```

```python
# Plot each neighbourhood's average YoY value difference in descending order on a bar chart.

plt.figure(figsize=(16, 9))

bars = plt.bar(
    x=df_diff_by_muni["Toronto Municipality"],
    height=df_diff_by_muni["avg_diff"])

# Add and format data labels.

for bar in bars:
    height = bar.get_height()
    label_value = height / 1000
    label_text = f"${label_value:,.0f}K"
    x_val = bar.get_x() + bar.get_width() / 2
    plt.text(
        x_val,
        height,
        label_text,
        ha="center",
        va="bottom")

# Add titles and axis formatting.

neighbourhood_diff_ax = plt.gca()
neighbourhood_diff_ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
plt.xticks(rotation=90)
plt.title("Average YoY Value Difference by Municipality (2015-2024)",fontsize=15)
plt.xlabel("Toronto Municipality", fontsize=15)
plt.ylabel("Average YoY Value Difference", fontsize=15)

plt.tight_layout()
plt.show()
```

#### Next, we can do a similar calcuation and bar chart to also look at the average percent difference YoY by neighbourhood.

```python
# Create a new data frame to group the YoY difference percentage by neighbourhood, property type and year.

pct_cols = [
    "HPI - Single-Family Detached YoY %",
    "HPI - Single-Family Attached YoY %",
    "HPI - Townhouse YoY %",
    "HPI - Apartment YoY %"
]

# Get the average YoY percentage difference for all property types by neighbourhood.

df_pct_by_muni = (
    df_final
    .groupby("Toronto Municipality")[pct_cols]
    .mean()               
    .mean(axis=1)
    .rename("avg_yoy_pct")
    .reset_index())

# Sort the data in descending order.

df_pct_by_muni.sort_values("avg_yoy_pct", ascending=False, inplace=True)
```

```python
# Plot each neighbourhood's average YoY percentage difference in descending order on a bar chart.

plt.figure(figsize=(16, 9))
bars = plt.bar(
    x=df_pct_by_muni["Toronto Municipality"], 
    height=df_pct_by_muni["avg_yoy_pct"])

# Add and format data labels.

for bar in bars:
    height = bar.get_height()
    label_text = f"{height:.1f}%"
    x_val = bar.get_x() + bar.get_width() / 2
    plt.text(
        x_val, 
        height, 
        label_text,
        ha="center", 
        va="bottom"
    )

# Add titles and axis formatting.

neighbourhood_diff_pct_ax = plt.gca()
neighbourhood_diff_pct_ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:.1f}%'))
plt.xticks(rotation=90)
plt.title("Average YoY Percentage Change by Municipality (2015-2024)",fontsize=15)
plt.xlabel("Toronto Municipality",fontsize=15)
plt.ylabel("Average YoY %",fontsize=15)

plt.tight_layout()
plt.show()
```
